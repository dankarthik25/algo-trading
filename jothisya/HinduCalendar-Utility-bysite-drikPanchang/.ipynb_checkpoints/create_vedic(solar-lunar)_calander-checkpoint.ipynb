```python
from datetime import datetime
from pytz import UTC # timezone
import pandas as pd
from icalendar import Calendar, Event
```

```python
def convert_ics2csv(file_path,panchanga_type):
    g = open(file_path,'rb')
    df = pd.DataFrame()
    gcal = Calendar.from_ical(g.read())
    # Panjan types[iskcon,telugu,kanada,tamil,marathi,oriya,bengali,gujarati,malayalam(kerala),assame,dainik,nepali]
    for component in gcal.walk():
        lunar_day = {}
        if component.name == "VEVENT":
            lunar_day['Date'] = component.get('dtstart').dt #'2021-01-01'
            lunar_day['Panchang'] = panchanga_type
            lunar_day['Location'] = component.get('location') #'Hyderabad,India'
            description = component.get('description')
            if 'Day Festivals and Events' in description:
                l1 =  description.split('Day Festivals and Events - \n')
                lunar_description = l1[0]
                lunar_festival = l1[1] #;    print(lunar_festival)
                festival_list = lunar_festival.split('\n')
                if '' in festival_list:
                    festival_list.remove('')
                if 'from www.drikpanchang.com' in festival_list:
                    festival_list.remove('from www.drikpanchang.com')    
                    lunar_day[panchanga_type +' Festivals'] = festival_list    
    #             print(festival_list) 
            else :
                lunar_description = description
            list=lunar_description.split('\n')
            for item in list:
                if '-' in item:
                    key_value=item.split('-')
                    lunar_day[key_value[0].strip()] = key_value[1].strip()
            df = df.append(lunar_day, ignore_index=True)                    
    #     print(lunar_day)
    # print(df)
    output_name = file_path.split('.')[0] + '.csv'
    df.to_csv(output_name)
    #         print("*Start Time*: ",component.get('dtstart').dt)       
    #         print('Location:',component.get('location'))
    #         print( "*Description*\n" , component.get('description'))
    #         print("*Summary*",component.get('summary'))
    #         print("*End Time*: ",component.get('dtend').dt)
    #         print("*DT STAMP*: ",component.get('dtstamp').dt)

    g.close()
```

Now create csv file for iskcon and telugu calendar and see the difference select which is field are required 

```python
# cal = Calendar.from_ical(open('DrikPanchang-2021.ics','rb').read())
file_path = 'DrikPanchang-iskcon-2021.ics'
# Panjan types[iskcon,telugu,kanada,tamil,marathi,oriya,bengali,gujarati,malayalam(kerala),assame,dainik,nepali]
panchanga_type = 'iskcon'
convert_ics2csv(file_path,panchanga_type)

file_path = 'DrikPanchang-telugu-2021.ics'
# Panjan types[iskcon,telugu,kanada,tamil,marathi,oriya,bengali,gujarati,malayalam(kerala),assame,dainik,nepali]
panchanga_type = 'telugu'
convert_ics2csv(file_path,panchanga_type)

# # # These are different parameter which are not present in vaishava 
# Telugu Festivals
# Lunar Month: There is a month differene in damodar masa and karthika masa 
# Samvatvara: 60 year Samvatvara cycle(https://sakalam.org/telugu-year-names-english/) 
# Karanamulu
# Rahu Kalam
# Gulikai Kalam
# Yamaganda
# Abhijit
# Dur Muhurtamulu
# Amrit Kalam
# Varjyam    

# df.drop(df.columns[[0,1,3]], axis=1, inplace=True)
# df_marks = df_marks.drop(['algebra', 'chemistry'], axis=1)
```
